```python
import numpy as np
from pylab import *
import waipy
z = np.linspace(0,2048,2048)
x = np.sin(50*np.pi*z)
y = np.cos(50*np.pi*z)
data_norm = waipy.normalize(x)
```

```python
result = waipy.cwt(data_norm, 1, 1, 0.125, 2, 4/0.125, 0.72, 6,mother='Morlet',name='x')
```

```python
waipy.wavelet_plot('Sine', z, data_norm, 0.03125, result)
```
